
# QuickBasket — A/B Test Analysis
# PMI Stage 1: Free Delivery on Second Order


## BUSINESS CONTEXT:
QuickBasket is a quick-commerce grocery delivery app operating  
in 8 Indian metros. Diagnostic analysis revealed that 60-75%  
of first-time buyers never return within Week 1, and the hazard  
rate for churn peaks at Day 14-18 after first order.

## EXPERIMENT:
We tested whether offering free delivery on the second order  
(nudge sent on Day 3, valid until Day 7) improves the 7-day  
return rate for first-time buyers.

## DESIGN:
- Control: Standard welcome discount, no structured follow-up  
- Treatment: Standard welcome discount + free delivery on 2nd order (Day 3-7)

##  DESIGN METRIC: 
- Primary metric: 7-day return rate  
- Sample: 1,181 per group (2,362 total)  
- MDE: 5 ppt (37% → 42%), α = 0.05, Power = 80%  
- Duration: 14 days (7 days enrollment + 7 days measurement)  

## GUARDRAILS:
- Second order AOV must not drop below ₹250  
- Cancellation rate must not exceed 8%  
- Discount per user must not exceed control by more than ₹40  

##

```python
import pandas as pd
import numpy as np
from scipy import stats
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:your_password@localhost:5432/ecom_tests')
```

```python
import numpy as np
from scipy.stats import norm

baseline = 0.37
alpha = 0.05
power = 0.80

z_alpha = norm.ppf(1 - alpha)  # one-sided test
z_beta = norm.ppf(power)

for mde in [0.03, 0.05, 0.07]:
    p1 = baseline
    p2 = baseline + mde
    p_avg = (p1 + p2) / 2
    
    numerator = (z_alpha * np.sqrt(2 * p_avg * (1 - p_avg)) + 
                 z_beta * np.sqrt(p1 * (1 - p1) + p2 * (1 - p2))) ** 2
    denominator = (p2 - p1) ** 2
    
    sample = int(np.ceil(numerator / denominator))
    daily_new_users = 10000 / 30
    days_needed = (sample * 2) / daily_new_users
    
    print(f"MDE: {mde:.0%} ({p1:.0%} → {p2:.0%})")
    print(f"  Sample per group: {sample:,}")
    print(f"  Total sample: {sample * 2:,}")
    print(f"  Days to collect: ~{days_needed:.0f} days")
    print()
```

## 02_Primary metric — 7-day return rate

```python
control = df[df['experiment_group'] == 'control']
treatment = df[df['experiment_group'] == 'treatment']
```

```python
control_returned = control['returned_7d'].sum()
treatment_returned = treatment['returned_7d'].sum()
control_n = len(control)
treatment_n = len(treatment)
```

```python
control_rate = control_returned / control_n
treatment_rate = treatment_returned / treatment_n
lift = treatment_rate - control_rate
relative_lift = lift / control_rate
```

```python
print("=== PRIMARY METRIC: 7-Day Return Rate ===\n")
print(f"  Control:   {control_rate:.1%} ({control_returned}/{control_n})")
print(f"  Treatment: {treatment_rate:.1%} ({treatment_returned}/{treatment_n})")
print(f"  Absolute lift: {lift:.1%}")
print(f"  Relative lift: {relative_lift:.1%}")
```

##

## 03_Statistical significance — z-test for proportions

```python
p_pool = (control_returned + treatment_returned) / (control_n + treatment_n)
se = np.sqrt(p_pool * (1 - p_pool) * (1/control_n + 1/treatment_n))
z_stat = lift / se
p_value = 1 - stats.norm.cdf(z_stat)
```

```python
print(f"\n--- Statistical test (one-sided z-test) ---")
print(f"  Pooled proportion: {p_pool:.4f}")
print(f"  Standard error: {se:.4f}")
print(f"  Z-statistic: {z_stat:.3f}")
print(f"  P-value: {p_value:.4f}")
print(f"  Significant at α=0.05: {'YES' if p_value < 0.05 else 'NO ✗'}")



# Confidence interval for the lift
se_lift = np.sqrt(
    control_rate * (1 - control_rate) / control_n +
    treatment_rate * (1 - treatment_rate) / treatment_n
)
ci_lower = lift - 1.96 * se_lift
ci_upper = lift + 1.96 * se_lift
print(f"\n  95% CI for absolute lift: [{ci_lower:.1%}, {ci_upper:.1%}]")
print(f"  Interpretation: The true lift is between {ci_lower:.1%} and {ci_upper:.1%}")

```

---------------------

##

## 04_ Guardrail metrics Check

### Only users who placed second order 

```python
control_returned = control[control['returned_7d'] == 1]
treatment_returned = treatment[treatment['returned_7d'] == 1]
```

#### Guardrail 1: Second order AOV

```python
c_aov = control_returned['second_order_value'].mean()
t_aov = treatment_returned['second_order_value'].mean()
t_stat, p_val = stats.ttest_ind(
    control_returned['second_order_value'].dropna(),
    treatment_returned['second_order_value'].dropna()
)
aov_pass = t_aov >= 250

print(f"Second Order AOV (threshold: ≥ ₹250)")
print(f"   Control:   ₹{c_aov:.0f}")
print(f"   Treatment: ₹{t_aov:.0f}")
print(f"   Difference: ₹{t_aov - c_aov:.0f}")
print(f"   p-value: {p_val:.4f} ({'Significant' if p_val < 0.05 else 'Not significant'})")
```

#### Guardrail 2: Cancellation rate

```python
c_cancel = (control_returned['second_order_status'] == 'Cancelled').mean()
t_cancel = (treatment_returned['second_order_status'] == 'Cancelled').mean()
cancel_pass = t_cancel <= 0.08

print(f"Cancellation Rate (threshold: ≤ 8%)")
print(f"   Control:   {c_cancel:.1%}")
print(f"   Treatment: {t_cancel:.1%}")
print(f"   Difference: {t_cancel - c_cancel:.1%}")
```

#### Guardrail 3: Discount per user in second order

```python
c_disc = control_returned['second_order_discount_value'].mean()
t_disc = treatment_returned['second_order_discount_value'].mean()
disc_diff = t_disc - c_disc
disc_pass = disc_diff <= 40

print(f"Second Order Discount (threshold: treatment ≤ control + ₹40)")
print(f"   Control:   ₹{c_disc:.1f}")
print(f"   Treatment: ₹{t_disc:.1f}")
print(f"   Difference: ₹{disc_diff:.1f}")
```

----------------------------------------

## 05_Secondary metric & segment analysis

```python
c_w2 = control['w2_returned'].mean()
t_w2 = treatment['w2_returned'].mean()
w2_lift = t_w2 - c_w2
```

#### Significance test for W2

```python
p_pool_w2 = df['w2_returned'].mean()
se_w2 = np.sqrt(p_pool_w2 * (1 - p_pool_w2) * (1/control_n + 1/treatment_n))
z_w2 = w2_lift / se_w2
p_w2 = 1 - stats.norm.cdf(z_w2)

print(f"  Control:   {c_w2:.1%}")
print(f"  Treatment: {t_w2:.1%}")
print(f"  Lift: {w2_lift:.1%}")
print(f"  P-value: {p_w2:.4f}")
print(f"  Significant: {'Yes' if p_w2 < 0.05 else 'No'}")
print(f"\n  Interpretation: {'Treatment effect sustains into W2 — genuine habit signal' if p_w2 < 0.05 else 'W2 effect not significant — intervention may be one-time'}")
```

#

#### SEGMENT ANALYSIS: Return Rate by Channel

```python
print(f"{'Channel':<20} {'Control':>10} {'Treatment':>10} {'Lift':>10} {'n_ctrl':>8} {'n_treat':>8}")
print("-" * 70)

for ch in sorted(df['acquisition_channel'].unique()):
    c_grp = control[control['acquisition_channel'] == ch]
    t_grp = treatment[treatment['acquisition_channel'] == ch]
    c_rate = c_grp['returned_7d'].mean()
    t_rate = t_grp['returned_7d'].mean()
    ch_lift = t_rate - c_rate
    print(f"{ch:<20} {c_rate:>9.1%} {t_rate:>9.1%} {ch_lift:>+9.1%} {len(c_grp):>8} {len(t_grp):>8}")
```

#

#### SEGMENT ANALYSIS: Return Rate by City

```python
print(f"{'City':<20} {'Control':>10} {'Treatment':>10} {'Lift':>10}")
print("-" * 55)

for city in sorted(df['city'].unique()):
    c_grp = control[control['city'] == city]
    t_grp = treatment[treatment['city'] == city]
    c_rate = c_grp['returned_7d'].mean()
    t_rate = t_grp['returned_7d'].mean()
    ch_lift = t_rate - c_rate
    print(f"{city:<20} {c_rate:>9.1%} {t_rate:>9.1%} {ch_lift:>+9.1%}")
```

## 06_Experiment Conclusion & Recommendation


### PRIMARY METRIC:
  7-day return rate improved from 35.8% to 42.2%
  Absolute lift: +6.4 ppts
  P-value: 0.0008 (significant at α = 0.05)
  95% CI: [+2.4%, +10.3%]

### GUARDRAILS:
  Second order AOV: ₹354 vs ₹364 — No significant drop (p=0.35)
  Cancellation rate: 3.6% vs 4.3% — No increase
  Discount usage: ₹2.4 vs ₹4.8 — Treatment used LESS discount

### SECONDARY METRIC:
  W2 return rate: 23.3% vs 18.7% — Significant (p=0.003)
  The treatment effect sustains beyond the incentive window.

### SEGMENT INSIGHT:
  Paid channel users (meta_ads, google_ads, influencer) showed 
  the largest lift (+6-9 ppts), confirming the diagnostic finding 
  that these users had the most room for improvement.

### PROJECTED IMPACT (monthly):
  Additional retained users: ~500-800 per month
  Revenue uplift: ₹8-13 lakh per month
  Intervention cost: ~₹1.2 lakh per month (free delivery waivers)
  Net monthly impact: ~₹7-12 lakh

### RECOMMENDATION:
  1. Ship the free delivery nudge to 100% of new users
  2. Monitor Pune city-level performance in scaled rollout
  3. Proceed to PMI Stage 2: test progressive incentives 
     for W2 and W3 retention to push users toward the 
     4-week habit lock-in threshold
